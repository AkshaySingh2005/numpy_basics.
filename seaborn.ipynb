**Seaborn is a Python data visualization library built on top of Matplotlib**

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rn=np.random.randint(50, 100, size=20)
an=np.arange(1,21)
print(rn)
print(an)
df = pd.DataFrame({"Rollno" : an , "Marks" : rn})
df
```

```python
sns.lineplot(x='Rollno',y="Marks",data=df)
plt.title("Student Marks")
```

```python
cd = pd.read_csv('/content/sample_data/california_housing_test.csv')
cd.head()
```

```python
sns.scatterplot(x='median_house_value',y='median_income',data=cd,hue='housing_median_age')
```

```python
sns.barplot(x='Rollno',y="Marks",data=df)
plt.title("Student Marks")
```

```python
df = pd.read_csv('/content/sample_data/california_housing_test.csv') #KDE is a smooth curve that estimates the probability density function (PDF) of a continuous variable.
sns.histplot(df['median_house_value'], bins=30, kde=True, color='skyblue') #bins are intervals (or "buckets") into which the range of data is divided.
plt.title('Distribution of Median House Value', fontsize=16)
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```
